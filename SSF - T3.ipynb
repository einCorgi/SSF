```python
tall_measure = []
tall_elements = []
counter = 100000

dset = open("C:\\Users\\bgdlr\\Documents\\SSF\\Datos A2-b\\altura6.txt")
dset_r = dset.read().splitlines()

total_of_elements = 0

for i in dset_r:
    tall_measure.append(float(i))
    total_of_elements += 1

average = sum(tall_measure)/total_of_elements

while counter > 0:
    for i in tall_measure:
        element = (i - average) ** 2
        tall_elements.append(element)
        counter -= 1

squared_sigma_variance = sum(tall_elements)/(total_of_elements)
sigma_standar_deviation = squared_sigma_variance**(1/2)
                                                   
print(f"\nEl promedio es: {average}")
print(f"La varianza es {squared_sigma_variance}")
print(f"La desviación estandar es: {sigma_standar_deviation}")
```
