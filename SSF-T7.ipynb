```python
import matplotlib.pyplot as plt
import numpy as np

h = 6.626e-34
c = 3.0e+8
k = 1.38e-23

def planck(wav, T):
    a = 8*np.pi*h*c
    b = h*c/(wav*k*T)
    intensity = a/ ( (wav**5) * (np.exp(b) - 1.0) )
    return intensity

# generate x-axis in increments from 1nm to 3 micrometer in 1 nm increments
# starting at 1 nm to avoid wav = 0, which would result in division by zero.
wavelengths = np.arange(1e-9, 3e-6, 1e-9) 

# intensity at 4000K, 5000K, 6000K, 7000K 4000
intensity4000 = planck(wavelengths, 4000.) # Red Star Temperature
intensity5000 = planck(wavelengths, 5000.) # White Star (Like our sun)
intensity6000 = planck(wavelengths, 6000.)
intensity7000 = planck(wavelengths, 7000.) # Blue Star


plt.plot(wavelengths, intensity4000, 'r-') # plot intensity 4000 versus wavelength in nm as a red line
plt.plot(wavelengths, intensity5000, 'g-') # 5000K green line
plt.plot(wavelengths, intensity6000, 'b-') # 6000K blue line
plt.plot(wavelengths, intensity7000, 'k-') # 7000K black line

# show the plot
plt.show()
```

```python
# Wien's Displacement Law
# ---> from 1e-9 to 3e-6
# ------> through 1e-9 steps per event

def wein(wav, T):
    cons = (h*c)/k
    x = -cons/(wav*T)
    BlackBodyPeak = cons/(5*(1-np.exp(x)))
    return BlackBodyPeak

# starting at 1 nm to avoid wav = 0, which would result in division by zero.
wavelengths = np.arange(1e-9, 3e-6, 1e-9) 

# intensity at 4000K, 5270K, 6000K, 7000K
BlackBodyPeak4000 = wein(wavelengths, 4000.) # Red Star Temperature
BlackBodyPeak5000 = wein(wavelengths, 5270.) # White Star (Like our sun)
BlackBodyPeak6000 = wein(wavelengths, 6000.)
BlackBodyPeak7000 = wein(wavelengths, 7000.) # Blue Star


plt.plot(wavelengths, BlackBodyPeak4000, 'r-') # plot intensity4000 versus wavelength in nm as a red line
plt.plot(wavelengths, BlackBodyPeak5000, 'g-') # 5270K green line
plt.plot(wavelengths, BlackBodyPeak6000, 'b-') # 6000K blue line
plt.plot(wavelengths, BlackBodyPeak7000, 'k-') # 7000K black line

# show the plot
plt.show()
```

So, We've numerically demonstrated that:
<h3><center> $\lambda T = \frac{a}{5[1-e^{-a/{\lambda T}}]} = 2.898 \times 10^{-3}$ m$\cdot$K </center></h3>

Technically, the $\lambda$ value is the $\lambda$ max value ($\lambda_{peak}$), so:
<h3><center> $\lambda_{peak} T = \frac{a}{5[1-e^{-a/{\lambda T}}]} = 2.898 \times 10^{-3}$ m$\cdot$K </center></h3>

```python
import matplotlib.pyplot as plt
import numpy as np

h = 6.626e-34
c = 3.0e+8
k = 1.38e-23

def Lambda_max(wav, T):
    a = 8*np.pi*h*c
    b = h*c/(wav*k*T)
    intensity = (h*c)/(wav*k*T)
    return intensity

# generate x-axis in increments from 1nm to 3 micrometer in 1 nm increments
# starting at 1 nm to avoid wav = 0, which would result in division by zero.
wavelengths = np.arange(1e-9, 3e-6, 1e-9) 

# intensity at 4000K, 5000K, 6000K, 7000K
intensity4000 = Lambda_max(wavelengths, 4000.) # Red Star Temperature
intensity5000 = Lambda_max(wavelengths, 5270.) # White Star (Like our sun)
intensity6000 = Lambda_max(wavelengths, 6000.)
intensity7000 = Lambda_max(wavelengths, 7000.) # Blue Star


plt.plot(wavelengths, intensity4000, 'r-') # plot intensity4000 versus wavelength in nm as a red line
plt.plot(wavelengths, intensity5000, 'g-') # 5270K green line
plt.plot(wavelengths, intensity6000, 'b-') # 6000K blue line
plt.plot(wavelengths, intensity7000, 'k-') # 7000K black line

# show the plot
plt.show()
```
